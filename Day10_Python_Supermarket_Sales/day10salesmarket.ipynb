```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\supermarket_sales.csv")
```

```python
print(df.head())
print(df.shape)
```

```python
print(df.isnull().sum().sum())
```

```python
print(df.columns.tolist())
```

```python
df.groupby('Product line')['gross income'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product line')['gross income'].sum().sort_values(ascending=False).plot(kind='bar',color='gray')
plt.title("Sum of gross income by product line")
plt.xlabel("product line")
plt.ylabel("gross income")
plt.show()
```

```python
df.groupby('City')['Total'].sum().sort_values(ascending=False).plot(kind='bar',color=['purple','gray','orange'])
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Total")
plt.show()
```
